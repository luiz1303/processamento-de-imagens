### 1) Utilize o elemento estruturante abaixo:




a) Aplique erosão e dilatação na imagem img1.png disponível no diretório. 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define o elemento estruturante
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))

img1 = cv2.imread('img1.png', cv2.IMREAD_GRAYSCALE)
assert img1 is not None, "Não foi possível ler o arquivo."

img1_erosion = cv2.erode(img1,kernel,iterations = 1)
img1_dilation = cv2.dilate(img1,kernel,iterations = 1)

plt.figure(figsize=(15, 5))
####
plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Imagem 1 Original')
####
plt.subplot(1, 3, 2)
plt.imshow(img1_erosion, cmap='gray')
plt.title('Imagem 1 com Erosão')
####
plt.subplot(1, 3, 3)
plt.imshow(img1_dilation, cmap='gray')
plt.title('Imagem 1 com Dilatação')
####
plt.show()
```

b) Aplique erosão, dilatação, fechamento na imagem img2.png

```python
img2 = cv2.imread('img2.png', cv2.IMREAD_GRAYSCALE)
assert img2 is not None, "Não foi possível ler o arquivo."

img2_erosion = cv2.erode(img2,kernel,iterations = 1)
img2_dilation = cv2.dilate(img2,kernel,iterations = 1)
img2_closing = cv2.morphologyEx(img2, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(15, 5))
####
plt.subplot(1, 4, 1)
plt.imshow(img2, cmap='gray')
plt.title('Imagem 2 Original')
####
plt.subplot(1, 4, 2)
plt.imshow(img2_erosion, cmap='gray')
plt.title('Imagem 2 com Erosão')
####
plt.subplot(1, 4, 3)
plt.imshow(img2_dilation, cmap='gray')
plt.title('Imagem 2 com Dilatação')
####
plt.subplot(1, 4, 4)
plt.imshow(img2_closing, cmap='gray')
plt.title('Imagem 2 com Fechamento')
####
plt.show()
```

### 2) Considere a imagem img4.png

a) Aplique o filtro morfológigo de erosão (3x3). Qual é o efeito na imagem?


```python
img4 = cv2.imread('img4.png', cv2.IMREAD_GRAYSCALE)
assert img4 is not None, "Não foi possível ler o arquivo."

kernelEx2 = np.ones((3,3),np.uint8)

plt.imshow(img4, cmap='gray')
plt.title('Imagem 4 Original')
```

```python
img4_erosion = cv2.erode(img4,kernelEx2,iterations = 1)
plt.imshow(img4_erosion, cmap='gray')
plt.title('Imagem 4 com Erosão')

# Como o processo de erosão, a imagem perdeu partes importantes.
# O último elemento chegou a desaparecer, causando um efeito indesejado.
```

b) Aplique o filtro morfológigo de dilatação (3x3). Qual é o efeito na imagem?

```python
img4_dilation = cv2.dilate(img4,kernelEx2,iterations = 1)
plt.imshow(img4_dilation, cmap='gray')
plt.title('Imagem 4 com Dilatação')

# A imagem com dilatação exibe menos ruído, visto que os pontos em vazio foram preenchidos
# com a imagem dilatada. Porém, a imagem perdeu detalhes do formato original.
```

c) Aplique o filtro morfológigo de abertura (3x3). Qual é o efeito na imagem?

```python
img4_opening = cv2.morphologyEx(img4, cv2.MORPH_OPEN, kernelEx2)
plt.imshow(img4_opening, cmap='gray')
plt.title('Imagem 4 com Abertura')

# Como o processo de abertura é uma erosão seguida de uma dilatação, a imagem perdeu partes importante
# devido a erosão. As partes restantes foram preenchidas com a dilatação.
```

d) Aplique o filtro morfológigo de fechamento (3x3). Qual é o efeito na imagem?

```python
img4_closing = cv2.morphologyEx(img4, cv2.MORPH_CLOSE, kernelEx2)
plt.imshow(img4_closing, cmap='gray')
plt.title('Imagem 4 com Fechamento')

# Como o Fechamento é uma dilatação seguida de uma erosão, os ruídos da imagem foram reduzídos
# devido a etapa inicial e, o formato foi expandido devido a segunda etapa.
```
