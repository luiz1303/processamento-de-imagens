1) Implemente o filtro de suavização pela média da vizinhança

```python
import numpy as np
import math
import cv2
import matplotlib.pyplot as plt

img1 = cv2.cvtColor(cv2.imread('img_aluno1.png'), cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(cv2.imread('img_aluno2.jpg'), cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12,12)) 

plt.subplot(121)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("Img 1")

plt.subplot(122)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title("Img 2")
```

2) Implemente o filtro de suavização pela média dos k vizinhos mais próximos

```python
k = 5;

w_size = 2*k+1;
kernel = np.ones((w_size, w_size), np.float32) / w_size**2
img_suavizada = cv2.filter2D(img2, -1, kernel)

plt.figure(figsize=(12,12)) 

plt.subplot(121)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB), cmap="gray", vmin=0, vmax=255)
plt.title("Img original")

plt.subplot(122)
plt.imshow(cv2.cvtColor(img_suavizada, cv2.COLOR_BGR2RGB), cmap="gray", vmin=0, vmax=255)
plt.title("Img com filtro")
```

3) Implemente o filtro da mediana.

```python
median=cv2.medianBlur(img2,15)

plt.figure(figsize=(12,12)) 

plt.subplot(121)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB), cmap="gray", vmin=0, vmax=255)
plt.title("Img original")

plt.subplot(122)
plt.imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB), cmap="gray", vmin=0, vmax=255)
plt.title("Img com filtro")
```

4) Implemente o operador Laplaciano.

```python
lapImg = cv2.Laplacian(img1,cv2.CV_64F)

plt.figure(figsize=(12,12)) 

plt.subplot(121)
plt.imshow(img1, cmap="gray", vmin=0, vmax=255)
plt.title("Original")

plt.subplot(122)
plt.imshow(lapImg, cmap="gray", vmin=0, vmax=255)
plt.title("Laplace")
```

5) Implemente o detector de bordas de Roberts.

```python
r_h1 = np.array([[1,0], [0,-1]])
r_h2 = np.array([[0,1], [-1,0]])

img_Rx = cv2.filter2D(img1, -1, r_h1)
img_Ry = cv2.filter2D(img1, -1, r_h2)
img_roberts = img_Rx + img_Ry

plt.figure(figsize=(12,12)) 

plt.subplot(121)
plt.imshow(img1, cmap="gray", vmin=0, vmax=255)
plt.title("Original")

plt.subplot(122)
plt.imshow(img_roberts, cmap="gray", vmin=0, vmax=255)
plt.title("Roberts")
```

6) Implemente o detector de bordas de Prewitt

```python
p_h1 = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
p_h2 = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

img_Px = cv2.filter2D(img1, -1, p_h1)
img_Py = cv2.filter2D(img1, -1, p_h2)
img_prewitt = img_Px + img_Py

plt.figure(figsize=(12,12))

plt.subplot(121)
plt.imshow(img1, cmap="gray", vmin=0, vmax=255)
plt.title("Original")

plt.subplot(122)
plt.imshow(img_prewitt, cmap="gray", vmin=0, vmax=255)
plt.title("Prewitt")
```

7) Implemente o detector de bordas de Sobel

```python
s_h1 = np.array([[1, -2, -1],[0,0,0],[1, 2, 1]])
s_h2 = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])

img_Sx = cv2.filter2D(img1, -1, s_h1)
img_Sy = cv2.filter2D(img1, -1, s_h2)
img_sobel = img_Sx + img_Sy

plt.figure(figsize=(12,12))

plt.subplot(121)
plt.imshow(img1, cmap="gray", vmin=0, vmax=255)
plt.title("Original")

plt.subplot(122)
plt.imshow(img_sobel, cmap="gray", vmin=0, vmax=255)
plt.title("Sobel")
```
